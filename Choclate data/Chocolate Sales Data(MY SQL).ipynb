```python
from pandas_profiling import ProfileReport as pdp
```

```python
#Display output in jupyter notebook
report = pdp(df_sales, title =' Sales Report')
report.to_widgets()
```

```python
# Minimal reports

report_minimal = pdp(df_sales, title =' Sales Report', minimal =True)
report_minimal.to_widgets()
report_minimal.to_file(output_file= 'Sales report minimal.html')
```
