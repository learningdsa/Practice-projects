#  3. First plot using plt.plot

```python
import numpy as np

x = np.array([1, 2, 6, 8])
y = np.array([3, 8, 1, 10])
y1 = np.array([5, 7, 10, 5])

fig = plt.figure()

plt.plot(x, y, '-')
plt.plot(x, y1, '-');
```

#   4. Saving to file 

```python
fig.savefig('my_figure.png')
```

# 5. Playing with Options 

```python
# For short, you can use the following codes:
plt.plot(x, y, linestyle='dashed', color='black', marker='o',markersize=12, linewidth=2)  # solid
```

#  6. Adjusting the Plot: Axes Limits

```python
plt.plot(x, np.sin(x))

plt.xlim(-1, 11)
plt.ylim(-1.5, 1.5);
```

# 7. Labeling Plots

```python
plt.plot(x, y)
plt.title("A Simple graph")
plt.plot(x, y, '-g', label='first graph')
plt.plot(x, y1, ':b', label='second graph')

plt.legend();
```

# 8. Scatter Plots

```python

plt.scatter(x, y, marker='o');
```

# 9. Histogram and Bins

```python
data = np.random.randn(1000)
```

```python
plt.hist(data);
```

```python
plt.hist(data, bins=10,
         histtype='stepfilled', color='steelblue',
         edgecolor='black');
```

# 10. Bar plot

```python
plt.bar(x, y, color='green')
```

# 11. Pie Chart 

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
mycolors = ["black", "pink", "yellow", "red"]

plt.pie(y, labels = mylabels, colors = mycolors)
plt.show() 
```

# 12. Combine plots 

```python

plt.plot(x, y, color= 'blue', lw=4, alpha=0.5)
plt.scatter(x, y, marker='o', s=400, color='red',alpha=0.5)
plt.bar(x, y, color= 'green',alpha=0.5)
```

 # 13. Sub Plots

```python
#plot 1:

plt.subplot(2, 2, 1)
plt.plot(x,y)

#plot 2:

plt.subplot(2, 2, 2)
plt.plot(x **2 ,y **2)


#plot 3:

plt.subplot(2, 2, 3)
plt.plot(x * 2,y * 2)


plt.show()
```
